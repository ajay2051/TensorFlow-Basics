Introduction to Tensors

What we have created so far:
* Scalar: a single number
* Vector: a number with direction(e.g wind speed with direction)
* Matrix: a 2-dimensional array of numbers
* Tensor: an n-dimensional array of numbers(when n can be any number, a 0-dimensional tensor is a scalar, a 1-dimensional tensor is a vector)

Creating random tensors

Random tensors are tensors of some arbitary size which contains random numbers.

```python
# Create two random (but the same) tensors
random_1 = tf.random.Generator.from_seed(42) # set seed for reproductibility
random_1 = random_1.normal(shape=(3,2))

random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))

random_1, random_2, random_1 == random_2
```

```python
# Schuffle the order of elements in a tensor
# Schuffle a tensor (valuable for when you want to schuffle data so the inherent order doesn't effect learning)

not_schuffled = tf.constant([[10, 7],
                             [3, 4],
                             [2, 5]
                             ])

# Schuffle our non_schuffled
shuffle = tf.random.shuffle(not_schuffled)
shuffle
```

```python
# shuffle our non-shuffled tensor
import tensorflow as tf
not_schuffled = tf.constant([[10, 7],
                             [3, 4],
                             [2, 5]
                             ])
tf.random.set_seed(42)  # Global level random seed
tf.random.shuffle(not_schuffled, seed=42)  # Operation level random seed


# It looks like if we want our suffled tensors to be in same order, we've to use the global level random seed as well as the operation level random seed.
# Rule: 4 If both the global and the operation seed are set: Both seeds are used in conjunction to determine the random sequence.
```

Turn NumPy arrays into tensors

The main difference between NumPy arrays and TensorFlow tensors is that tensors can be run on GPU much faster computing.

```python
import numpy as np
numpy_A = np.arange(1, 25, dtype=np.int32)  # Range (1, 25) generates 24 number
tensor_a = tf.constant(numpy_A)
tensor_b = tf.constant(numpy_A, shape=(4 ,3, 2)) # While reshaping the product should be 24 just like 2 * 3 * 4 = 24
tensor_a
tensor_b
```

Getting Information from tensors
When dealing with tensors you probably want to be aware of the following attributes:
* Shape
* Rank
* Axis or dimension
* Size

```python
# Create rank 4 tensor (4 dimensions)

rank_4_tensor = tf.zeros(shape=[2, 3, 4, 5])
rank_4_tensor
rank_4_tensor.ndim
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape, tf.size(rank_4_tensor)
```

```python
# get various attributes of our tensor
print("Data Type of every element:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensors:", rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("Elements along the last axis:", rank_4_tensor.shape[-1])
print("Total number of elements in our tensor:", tf.size(rank_4_tensor))
```

Indexing Tensors

Tensors can be indexed just like Python lists.

```python
# Get the first two elements of each dimension
rank_4_tensor[:2, :2, :2, :2]
```

```python
# Get the first element from each element from each dimension from each index except for the final one
rank_4_tensor[:1, :1, :1, :]
```

```python
# Create a rank 2 tensor( 2 dimensions )
rank_2_tensor = tf.constant([[10, 7], [3, 4]])
rank_2_tensor.shape, rank_2_tensor.ndim
```

```python
# Get the last item of each of row of our 2 tensor
rank_2_tensor[:, -1]
```

```python
# Add in extra dimension to our rank 2 tensor

rank_3_tensor = rank_2_tensor[..., tf.newaxis]  # newaxis add in extra dimension
rank_3_tensor
```

```python
# Alternative to tf.newaxis

tf.expand_dims(rank_2_tensor, axis=-1)  # -1 means expand the final axis
```

```python
tf.expand_dims(rank_2_tensor, axis=0)  # expand the 0-axis
```

Manipulating tensors (tensors operations)

Basic Operations
`+` , `-` , `*`, `/`

```python
# You can add values to a tensor using the addition operator

tensor = tf.constant([[10, 7], [3, 4]])
tensor + 10
```

```python
# Original tensor is unchanged
tensor
```

```python
# Multiplication also works

tensor * 10
```

```python
# Subtraction also works

tensor - 10
```

```python
# We can use the tensorflow built-in function too

tf.multiply(tensor, 10)
```

Matrix Multiplication

In Machine Learning, matrix multiliplication is one of the most common tensor operations.

```python
# Matrix multiplication in tensorflow
a = tf.constant([[1,2,7],[7,2,1],[3,3,3,]])
b = tf.constant([[2,5],[6,7],[1,8]])
tf.matmul(a, b)
```

```python
# Matrix multiplication with Python Operator "@"

c = a @ b
```
