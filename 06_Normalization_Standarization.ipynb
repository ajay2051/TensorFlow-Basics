#Preprocessing and Normalization

#Preprocessing data (normalization and standardization)

Normalization is scaling the data to be analyzed to a specific range such as [0.0, 1.0] to provide better results.

In terms of scaling values, neural networks tend to prefer normalization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
salary_data = pd.read_csv("/content/salaries.csv")
salary_data
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
# Create a column transformer

ct = make_column_transformer((MinMaxScaler(), ["job_title","salary"]),
                             (OneHotEncoder(handle_unknown='ignore'), ["salary_in_usd","employee_residence"])) # turn all values in these columns 0 and 1
```

```python
# Create x and y

x = salary_data.drop("job_title", axis=1)
y = salary_data["salary"]


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
len(x), len(x_train), len(x_test), len(y_train), len(y_test)
```

```python
# Fit the column transformer to our training data

ct.fit(x_train)
```

```python
# Transform training and test data with normalization (MinMaxScaler) and OneHotEncoder
x_train_normal = ct.transform(x_train)
x_test_normal = ct.transform(x_test)
```

```python
# what does our data look like?

x_train.loc[0]
```

```python
# Build a neural network model to fit on our normalalized data

# set Random Seed
tf.random.set_seed(42)

# 1. Create the model

salaries_model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])


# 2. Compile the Model
salaries_model_3.compile(
    loss=tf.keras.losses.mae,
    optimizer=tf.keras.optimizers.Adam(),# SGD was not working
    metrics=["mae"]
)

# 3. Fit the model
history = salaries_model_3.fit(x_train_normal, y_train, epochs=200, verbose=1)
```
